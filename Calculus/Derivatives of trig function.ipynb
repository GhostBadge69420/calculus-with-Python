# Exercise

fa(x) = cos(x + sin(x)) + a                                     

a ∈{0,1,2,3}

plot these function in same graph

plot their derivatives in same graph(but different from the original functions)

```python
x,a = sym.symbols('x,a')

f = sym.cos(x + sym.sin(x)) + a

for ai in range(0,4):
    if ai==0:
        p = symplot(f.subs(a,ai),show=False,label='a=' + str(ai))
    else:
        p.extend(symplot(f.subs(a,ai),show=False,label='a=' + str(ai)))

p.show()
```

```python
x,a = sym.symbols('x,a')

f = sym.cos(x + sym.sin(x)) + a
colors = 'brkm'

for ai in range(0,4):
    if ai==0:
        p = symplot(f.subs(a,ai),show=False,label='a=' + str(ai),line_color=colors[0])
    else:
        p.extend(symplot(f.subs(a,ai),show=False,label='a=' + str(ai),line_color=colors[ai]))

p.show()
```

```python
x,a = sym.symbols('x,a')

f = sym.cos(x + sym.sin(x)) + a
colors = 'brkm'

for ai in range(0,4):
    if ai==0:
        p = symplot(f.subs(a,ai),show=False,label='a=' + str(ai),line_color=colors[0])
    else:
        p.extend(symplot(f.subs(a,ai),show=False,label='a=' + str(ai),line_color=colors[ai]))

p.title = 'The functions'
p.legend = True
p.show()
```

```python
x,a = sym.symbols('x,a')

f = sym.cos(x + sym.sin(x)) + a
colors = 'brkm'

for ai in range(0,4):
    if ai==0:
        p = symplot(f.subs(a,ai),show=False,label='a=' + str(ai),line_color=colors[0])
    else:
        p.extend(symplot(f.subs(a,ai),show=False,label='a=' + str(ai),line_color=colors[ai]))

p.title = 'The functions'
p.legend = True
p.show()


for ai in range(0,4):
    if ai==0:
        p = symplot(sym.diff(f.subs(a,ai)),show=False,label='a=' + str(ai),line_color=colors[0])
    else:
        p.extend(symplot(sym.diff(f.subs(a,ai)),show=False,label='a=' + str(ai),line_color=colors[ai]))

p.title = 'Their derivatives'
p.legend = True
p.show()
```
